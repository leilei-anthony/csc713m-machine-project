# **Section 2. Description of the dataset**
*This section provides details about the dataset, including its source, structure, and key attributes.*

### Dataset Overview
The **Family Income and Expenditure Survey (FIES) 2012** is a nationally representative survey conducted by the Philippine Statistics Authority (PSA) to collect detailed information on **household income**, **expenditure**, and other **socioeconomic indicators**. It provides insights into the financial conditions of Filipino households, allowing analysis of income distribution, expenditure patterns, and economic well-being. The dataset includes various income sources such as wages, business profits, remittances, and government support, as well as household expenditures on food, housing, education, and other essentials.

### Data Collection Process
The FIES 2012 dataset was collected through a **nationwide household survey** conducted by the PSA. Data collection involved structured interviews using survey questionnaires, where household heads provided responses on income sources, expenditures, and demographic details. The data was gathered through stratified sampling, ensuring representation across different regions, urban and rural areas, and various income brackets.

One limitation of the data collection process is its **reliance on self-reported income and expenditure**, which may introduce reporting biases or inaccuracies due to misestimation or underreporting of income. Additionally, economic conditions in 2012 may differ significantly from present-day scenarios, limiting the dataset's applicability to current financial trends. However, despite these limitations, the dataset remains valuable for understanding historical income distribution and economic trends in the Philippines.

### Structure of the Dataset
The dataset is stored as a CSV file named "FIES PUF 2012 Vol.1.CSV", which contains structured tabular data. It includes multiple columns representing household characteristics, income sources, expenditures, and demographic factors.

# **Section 6. Initial model training**
*This section covers the first stage of training a model using the prepared dataset, including the selection of algorithms, training process, and initial performance evaluation.*

### Evaluation Metrics

1. **Mean Squared Error (MSE)**  
  - MSE measures the average squared difference between the actual values (**y_test**) and the predicted values (**y_pred**).  
  - It tells you how far your predictions are from the actual values on average.  
  - A lower MSE indicates better model performance. However, since MSE is in squared units, it can be sensitive to outliers.  

2. **Root Mean Squared Error (RMSE)**  
  - RMSE is the square root of MSE, making it easier to interpret since it has closer values to the target variable.  
  - It measures how much, on average, predictions deviate from actual values.  
  - A lower RMSE indicates better model performance, but like MSE, it is sensitive to large errors (outliers).  

3. **Mean Absolute Error (MAE)**  
  - MAE measures the average absolute difference between actual and predicted values.  
  - Unlike MSE and RMSE, MAE is **less sensitive to outliers** because it does not square the errors.  
  - A lower MAE means predictions are consistently close to the actual values.  

4. **R-squared (R²) Score**  
  - R² measures the proportion of variance in the target variable that is explained by the features.  
  - It shows how well the model fits the data.  
    - **R² = 1:** Perfect fit (all data points are perfectly predicted).  
    - **R² = 0:** The model is no better than predicting the mean of **y**.  
    - **R² < 0:** The model performs worse than predicting the mean.

### Neural Networks (TOINC)


Neural network regression is a powerful non-linear model that learns patterns in data by passing inputs through multiple layers of interconnected neurons. Each neuron applies activation functions to transform inputs, enabling the model to capture complex relationships between features and the target variable. The model adjusts its weights during training to minimize error, making it highly adaptable for predictive tasks.  

It is suitable for the dataset because:  
- **Capturing Complex Patterns**: Neural networks can model intricate, non-linear relationships that traditional regression methods may struggle with.  
- **Scalability with Large Data**: They perform well with large datasets, learning from vast amounts of information to improve accuracy.  
- **Robust Feature Interactions**: Unlike traditional models, neural networks can automatically learn interactions between features without requiring manual feature engineering.

Step 2: Define the neural network model and its layers

Gets the number of features needed for the network

```python
num_features = X_train.shape[1]
```

Sets the number of hidden layers proportional to the number of features as a default

```python
num_hidden_layers = max(2, int(np.log2(num_features)))
```

Instantiates the neural network model and sets the number of inputs in the input layer to the number of features.

```python
nn_model = Sequential()
nn_model.add(Input(shape=(num_features,)))
```

Adds hidden layers equal to the amount that as specified to the network with an activation using ReLu

```python
for i in range(num_hidden_layers):
    units = max(16, num_features // (2 ** (i + 1)))
    nn_model.add(Dense(units, activation='relu'))
    nn_model.add(Dropout(0.2))
```

Appends an output layer to the neural network that outputs a linear value

```python
nn_model.add(Dense(1, activation='linear'))
```

Step 3: Compile the model

```python
nn_model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

Step 4: Train the Model

```python
history = nn_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=32)
```

```python
nn_y_train_pred = nn_model.predict(X_train)
```

Hyperparameters for Decision Trees

## Neural Network Hyperparameter Tuning Using RandomSearch

Hyperparameters for Neural Networks

* **Number of Hidden Layers (`num_hidden_layers`)** - The number of layers between the input and output layers (from 2 - 6).
* **Number of Neurons per Layer (`units_X`)** -  The number of neurons (units) in each hidden layer (from 16 - 128).
* **Activation Function** - Defines how neurons process the weighted sum of inputs.

Regularization Hyperparameters

* **Dropout Rate (`dropout_X`)** - The fraction of neurons randomly dropped during training to prevent overfitting (from 0.1 - 0.5).

Optimization Hyperparameters

* **Learning Rate (`learning_rate`)** - Determines how much weights update during each step of training.

Step 1: Define the parameter grid and model for Neural Network Grid Search

```python
def build_model(hp):
    op_nn_model = Sequential()

    # Input Layer
    op_nn_model.add(Input(shape=(X_train.shape[1],)))

    for i in range(hp.Int('num_hidden_layers', min_value=2, max_value=6)): # 2 and 6 layers
        units = hp.Int(f'units_{i}', min_value=16, max_value=128, step=16)  # 16-128 neurons per layer
        op_nn_model.add(Dense(units, activation='relu'))
        dropout_rate = hp.Float(f'dropout_{i}', min_value=0.1, max_value=0.5, step=0.1)  # Dropout range 0.1-0.5
        op_nn_model.add(Dropout(dropout_rate))

    # Output Layer
    op_nn_model.add(Dense(1, activation='linear'))

    # Compile model with tunable learning rate
    learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])  # Learning rate choices
    op_nn_model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mse', metrics=['mse'])

    return op_nn_model
```

Step 2: Define the keras tuner that will be used to tune the hyperparamters of the Neural Network

```python
tuner = kt.RandomSearch(
    build_model,
    objective='mse',
    max_trials=10,
    executions_per_trial=1,
    directory='tuner_results',
    project_name='nn_tuning',
    overwrite=True
)
```

Step 3: Perform Hyperparameter testing on the neural network using the defined keras tuner

```python
tuner.search(X_train, y_train, epochs=50, validation_data=(X_val, y_val), verbose=1)
```

Step 4: Get the best model found by the tuner

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

Step 5: Print the best hyperparamters

```python
print("Best Hyperparameters:")
print(f"Number of Hidden Layers: {best_hps.get('num_hidden_layers')}")
for i in range(best_hps.get('num_hidden_layers')):
    print(f"Layer {i+1}: {best_hps.get(f'units_{i}')} units, Dropout {best_hps.get(f'dropout_{i}')}")

print(f"Learning Rate: {best_hps.get('learning_rate')}")
```
